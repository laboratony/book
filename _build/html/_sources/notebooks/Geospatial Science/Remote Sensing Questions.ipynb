11. For an Earth observation satellite on a circular, near-polar, sun-synchronous orbit, which of the statements below is correct?
    1. It can image the entire Earth's surface including the regions of the North and South poles.
    2. It can take images of near infrared spectral bands on both descending and ascending passes.
    3. It cannot take radar images on both descending and ascending passes.
    4. <span style="color: #35B957">It images different areas of the Earth’s surface at about the same local time.</span>
    
```{note}
- Small areas of North and South poles are not covered due to the inclination angle.
- NIR is not TIR.
- "sun-synchronous" means covering most of the earth (except the near polar areas), at nearly the same local time imaging.
```

12. Which statement about image geometric distortion below is <b>wrong</b>?
    1. Geometric distortion caused by Earth curvature is unavoidable to all satellite images.
    2. Push-broom scanning introduces less geometric distortion than across-track scanning.
    3. <span style="color: #35B957">For across-track scanners, one-way scanning introduces more geometric distortion than two-way scanning.</span>
    4. Air-borne remote sensing is not subject to the geometric distortion resulted from Earth rotation.

```{note}
This effect is typical for a push-broom scanner. For an across-track two way scanner (e.g. Thematic Mapper), the
distortion pattern is even more complex as every pixel along a swath is imaged at a slightly different time and the
actual scanning speed on the earth surface changes not only from nadir to the edge of a swath but also between
swathes for and against earth rotation.
```
